```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import platform

#   
if platform.system() == "Windows":
    rc("font", family="Malgun Gothic")
elif platform.system() == "Darwin":  # macOS
    rc("font", family="AppleGothic")
plt.rcParams["axes.unicode_minus"] = False  #    

#    
elementary_heights = []
middle_heights = []
high_heights = []

#   
with open(".csv", "r", encoding="euc-kr") as f:
    data = csv.reader(f)
    
    #    
    for row in data:
        grade = row[1]  #  
        height = float(row[2])  #  
        
        if grade == '':
            elementary_heights.append(height)
        elif grade == '':
            middle_heights.append(height)
        elif grade == '':
            high_heights.append(height)

#     
elementary_avg = sum(elementary_heights) / len(elementary_heights) if elementary_heights else 0
middle_avg = sum(middle_heights) / len(middle_heights) if middle_heights else 0
high_avg = sum(high_heights) / len(high_heights) if high_heights else 0

#  
print("  :", elementary_avg)
print("  :", middle_avg)
print("  :", high_avg)

#  
categories = ["", "", ""]
averages = [elementary_avg, middle_avg, high_avg]

plt.figure(figsize=(8, 5))
plt.bar(categories, averages, color=["skyblue", "lightgreen", "salmon"])
plt.xlabel("")
plt.ylabel("  (cm)")
plt.title(", ,   ")
plt.show()
```
